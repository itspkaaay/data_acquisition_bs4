# LOADING HTML

```python
from urllib.request import urlopen
```

```python
pl_url= 'https://en.wikipedia.org/wiki/2008%E2%80%9309_Premier_League'
pl= urlopen(pl_url)
print(type(pl))
```

```python

pl_html=(pl.read())
type(pl_html)
pl_html
```

# Parsing the HTML

```python
from bs4 import BeautifulSoup as soup
```

```python
pl_soup=soup (pl_html,'html.parser')
pl_soup
```

# EXTRACTING THE TABLE

```python
tables_list=pl_soup.findAll('table',{'class':'wikitable'})
print(len(tables_list))
print(tables_list)
tables_list[0]
```

 # Getting Column data

```python
list_headers=tables_list[0].findAll('th',{})
header=[x.text[:-1] for x in list_headers]
header
```

# Getting Row Data

```python
row_list=tables_list[0].findAll('tr',{})[1:]   #Slicing done to remove the column header as its taken care of
print(len(row_list))
```

```python
curr_row=row_list[0].findAll('td')
curr_row=[x.text[:-1] for x in curr_row]
curr_row
```

```python
listfinal=[]
listfinal.append(curr_row)

for row in row_list[2:]:
    curr_row=row.findAll('td')
    curr_row=[x.text[:-1] for x in curr_row]
    listfinal.append(curr_row)
    
listfinal
```

# WRITING ACQUIRED DATA INTO A COMMA SEPARATED FORM 

```python
with open('PLTeamInfo.csv','w') as file:
    headerstring=','.join(header)
    print(headerstring)
    headerstring+='\n'
    file.write(headerstring)
    
    for row in listfinal:
        row_info=''
        for w in row:
            w= w.replace(',','')
            row_info+=w
            row_info+=','
        row_info+='\n'
        print(row_info)
        file.write(row_info)
```
